```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Expanded_data_with_more_features.csv")
print(df.head())
```

```python
df.describe()
```

```python
#how to find not null values
df.info()
```

```python
#find null values
df.isnull().sum()
```

```python
#drop unnames column
df=df.drop("Unnamed: 0" , axis=1)
print(df.head())
```

```python
#change weekly study hours column
df["WklyStudyHours"]= df["WklyStudyHours"].str.replace("5 - 10","5-10")
df.head()
```

```python
#gender distribution 
plt.figure(figsize= (5,5))
ax= sns.countplot(data=df, x="Gender")
ax.bar_label(ax.containers[0])
plt.title("General Distribution")
plt.show()
```

```python
#form the above chart we have analysed that:
#the number of females in the data is more then the number of males
```

```python
gb= df.groupby("ParentEduc").agg({"MathScore":'mean', "ReadingScore":'mean', "WritingScore":'mean'})
print(gb)
```

```python
sns.heatmap(gb)
plt.show()
```

```python
#from the above chart we have concluded that the education of the parents have a good impact on  their scores
```

```python
gb1= df.groupby("ParentMaritalStatus").agg({"MathScore":'mean', "ReadingScore":'mean', "WritingScore":'mean'})
print(gb1)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb1, annot=True)
plt.title("Relationship between parent's Education and Students's Score")
plt.show()
```

```python
#from the above chart we have concludec that there is no/negligible impact on the
#student's score due to theirm parent's marital status
```
